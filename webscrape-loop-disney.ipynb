```python
# Import dependencies
from selenium import webdriver
from splinter import Browser
from bs4 import BeautifulSoup as soup
import matplotlib.pyplot as plt
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#kind of had to "brute force" the path
executable_path = "C:/Program Files/chromedriver-win64/chromedriver-win64/chromedriver.exe"
browser = Browser('chrome', executable_path=executable_path)
```

```python
# Visit the 1995
url = 'https://www.the-numbers.com/market/1995/distributor/Walt-Disney'
browser.visit(url)

# Parse the website
html = browser.html
```

```python
soup
```

```python
# Create a function to perform the web scraping

soup = BeautifulSoup(html, 'html.parser')

# Collect the HTML from the browser
html = browser.html
# Find the specific div with the 'page_filling_chart' ID
div = soup.find_all('div', id='page_filling_chart')
second_div = div[1]  # Access the second div

# Find the table within the second div
table = second_div.find('table')

# Extract data from the table, including the header
header_row = table.find('tr')
header = [cell.get_text(strip=True) for cell in header_row.find_all(['th', 'td'])]

data = []
for row in table.find_all('tr')[1:]:
    row_data = [cell.get_text(strip=True) for cell in row.find_all(['th', 'td'])]
    if row_data:
        data.append(row_data)

```

```python
pd.DataFrame(data, columns = header)
```
